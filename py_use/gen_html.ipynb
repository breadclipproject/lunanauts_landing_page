```python
traits_list = os.listdir('../traitsTable')
traits_list
```

```python
html_code = """
<div>
<table>
<tr>
    <td>{0}</td>
        <td>
            <div class="row">    
                <img class="traitsimg" src="{1}">
                <img class="traitsimg" src="{2}">
                <img class="traitsimg" src="{3}">
            </div>
   </td>
</tr>
</table> 
</div>
       
"""

htmlcode_formatted_combined = ""

for trait in traits_list:
    trait_dir = "..\\traitsTable" + "\\" + trait
    y = [x[0] for x in os.walk(trait_dir)]
    y.remove(trait_dir)
    
    for trait_name in y:
        #print(trait_name)
        x = os.listdir(trait_name)
   
        
        
        list_traits_dir = []
        
        for z in x:
            j = trait_name + "\\" + z
            list_traits_dir.append(j)
            
        word = list_traits_dir[0][:-10]
        the_trait_name = word.split('\\')[-1]
     
        htmlcode_formatted = html_code.format(the_trait_name, list_traits_dir[0][1:], list_traits_dir[1][1:], list_traits_dir[2][1:])
        
        print(htmlcode_formatted)
        
        #list_traits_dir = []
    htmlcode_formatted_combined = htmlcode_formatted_combined + htmlcode_formatted
```
